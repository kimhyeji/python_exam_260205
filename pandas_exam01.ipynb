```python
import pandas as pd

# 1. 아래와 같은 DataFrame을 생성하세요.
import pandas as pd

# 1. 아래와 같은 DataFrame을 생성하세요.
df = pd.DataFrame({
    "Name": ["Tom", "Jane", "Mike", "Sara", "John", "Emma"],
    "Age": [20, 22, 19, 21, 23, 20],
    "Score": [85, 92, 78, 88, 95, 70],
    "Grade": ["B", "A", "C", "B", "A", "C"]
})


# 2. 기본 정보 확인 -----------------------------------
# (1) DataFrame 전체 출력
print(df)

# (2) shape 출력
print(df.shape)

# (3) info() 출력
print(df.info())

# (4) describe() 출력
print(df.describe())


# 3. 열 선택 -----------------------------------
# (1) "Score" 열만 출력
print(df["Score"])

# (2) "Name"과 "Age" 열만 출력
print(df[["Name", "Age"]])


# 4. 행 선택 -----------------------------------
# (1) 첫 번째 행 출력 (iloc 사용)
print(df.iloc[0])

# (2) 마지막 행 출력 (iloc 사용)
print(df.iloc[-1])

# (3) index를 다음으로 변경
df.index = ["학생1", "학생2", "학생3", "학생4", "학생5", "학생6"]

# (4) "학생3" 행 출력 (loc 사용)
print(df.loc["학생3"])


# 5. 통계 계산 -----------------------------------
# (1) Score의 최대값
print("최대:", df["Score"].max())

# (2) Score의 최소값
print("최소:", df["Score"].min())

# (3) Score의 평균
print("평균:", df["Score"].mean())


# 6. 조건 선택 -----------------------------------
# (1) Score가 85 이상인 학생만 출력
print(df[df["Score"] >= 85])

# (2) Grade가 "A"인 학생만 출력
print(df[df["Grade"] == "A"])

# (3) Score가 평균보다 높은 학생만 출력
print(df[df["Score"] > df["Score"].mean()])


# 7. 일부만 보기 -----------------------------------
# (1) 상위 3개 출력
print(df.head(3))

# (2) 하위 2개 출력
print(df.tail(2))
```
