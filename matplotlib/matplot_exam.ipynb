```python
# matplotlib = 데이터를 표현할 수 있게 해주는 파이썬 라이브러리
# figure => 큰 도화지
# subplots => 도화지 위에 나눠진 작은 칸들
```

```python
import matplotlib.pyplot as plt

# 기본 스타일 적용
plt.style.use('default')

# y값만 전달, x축은 자동으로 0, 1, 2, 3이 설정
# plt.plot([0, 1, 2, 3], [2, 4, 6, 8])
plt.plot([2, 4, 6, 8])

plt.show()
```

```python
x = [2, 4, 6, 8]
y = [1, 3, 5, 7]
plt.plot(x, y)
plt.show()
```

```python
x = range(11) # 0 ~ 10
y = range(5, 16) # 5 ~ 15

# y : color, * : marker
# plt.plot(x, y, "mv")
plt.plot(x, y, "y*")
plt.show()
```

```python
fig = plt.figure() # 큰 도화지를  만듦

# 2 : 도화지를 2개의 행(row)으로 나눈다.
# 1 : 한 행에 1개의 열(column)이 있다
# 1 : 그 중에서 첫 번쨰 칸을 사용한다.(위쪽 칸)
fg1 = fig.add_subplot(2, 1, 1)
fg2 = fig.add_subplot(2, 1, 2)

fg1.plot([1, 2, 3], [4, 5, 6])
fg2.plot([2, 4, 6], [1, 3, 7])

plt.show()
```

```python
# 문제 1
# 1. 도화지를 생성한다.
# 2. 도화지를 가로 방향으로 3개의 칸으로 나눈다(1행 3열)
# 3. 그 중 첫 번째 칸과 세번째 칸에만 그래프 영역을 만든다.
#    - 두 번째 칸은 비워둔다

fig = plt.figure()

# 1행 3열 중 첫 번째 칸
ax1 = fig.add_subplot(1, 3, 1)

# 1행 3열 중 세 번째 칸
ax3 = fig.add_subplot(1, 3, 3)

ax1.plot([1, 2, 3], [1, 4, 9])
ax3.plot([1, 2, 3], [3, 2, 1])

plt.show()
```

```python
# 문제 2
# 1. 하나의 도화지를 생성한다.
# 2. 도화지를 1행 2열로 나눈 뒤, 첫번 째 칸에 그래프 영역을 하나 생성한다.
# 3. 같은 도화지를 다시 2행 2열로 나눈 뒤,
#    첫 번째 행의 두번째 칸에 그래프 영역을 하나 더 생성한다.

fig = plt.figure()

# 도화지를 1행 2열로 나누고, 첫 번째 칸에 그래프 영역 생성
fig.add_subplot(1, 2, 1)

# 도화지를 2행 2열로 나누고, 두 번째 칸(첫 번째 행, 두 번째 열) 그래프 영역 생성
fig.add_subplot(2, 2, 2)

plt.show()
```

```python
fig = plt.figure()

# 가로로 2칸(1행 2열), 첫 번째 칸
a1 = fig.add_subplot(1, 2, 1)

# 가로로 2칸(1행 2열), 두 번째 칸
a2 = fig.add_subplot(1, 2, 2)

x = range(0, 10)
y = range(5, 15)
# a1에 빨간색, 삼각뿔 아래쪽 모양으로 그래프
a1.plot(x, y, "r1")

x = range(0, 10)
y = range(5, 15)
# a2에 청록색, 육각형 모양으로 그래프
a2.plot(x, y, "ch")
```

```python
fig = plt.figure()

# 그래프 전체 디자인(테마) 변경
plt.style.use('ggplot')

# 도화지를 1행 1열로 나눠서, 한 칸(전체)에 그래프 영역
a = fig.add_subplot(111)

x = [3, 4] # 막대가 서 있는 위치(가로 위치)
y = [1, 2] # 각 막대의 높이(세로 값)

labels = ['english', 'math']
colors = ['#9eaeff', '#fff78a']

a.bar(x, y, tick_label=labels, color=colors)

plt.show()
```

```python
# 문제 3
# 1. 막대 그래프를 가로 방향으로 그린다.(힌트 : barh)
# 2. 막대의 값은 1, 2이고, 위치는 3, 4이다.
# 3. 각 막대의 값은 각각 원하는 색상으로 지정.
# 4. 라벨은 english, math

fig = plt.figure()

plt.style.use('ggplot')

a = fig.add_subplot(111)
x = [3, 4] # 막대의 위치(좌표 기준)
y = [1, 2] # 막대의 높이(값 기준)

labels = ['english', 'math']
colors = ['pink', 'black']

# bar가 아니라 barh를 쓰면 가로 막대를 그릴 수 있다.
a.barh(x, y, tick_label=labels, color=colors)
plt.show()
```

```python
# 막대 그래프 수치 표현
fig = plt.figure()

plt.style.use('ggplot')
a = fig.add_subplot(111)
x = [3, 4]
y = [1, 2]

labels = ['english', 'math']
colors = ['pink', 'black']

a.bar(x, y, tick_label=labels, color=colors)

for i, v in enumerate(y):
    # a.text(x좌표, y좌표, '보여줄 글자')
    a.text(v+2, i + 1.05, str(v)) # 표준 수치
    
plt.show()
```
