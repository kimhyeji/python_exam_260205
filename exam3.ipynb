```python
# 운영체제 모듈 불러오기
import os

# 현재 디렉토리를 알려주는 함수
print(os.getcwd())
```

```python
# 디렉토리를 이동할 수 있게 해주는 함수
# 한국어 인식이 안되기 때문에 웬만하면 영어로.
os.chdir("C://tmp")
print(os.getcwd())
os.chdir("C://work")
print(os.getcwd())
```

```python
# 입력한 경로의 파일과 폴더 목록을 리스트로 반환하는 함수
os.listdir("C://work")

# 리스트에 있는 요소 하나씩 출력
for i in os.listdir("C://work"):
    print(i)
```

```python
# 특정 폴더가 존재하는지 안 하는지 판단해주는 함수
print(os.path.exists("C://work/py_test"))
print(os.path.exists("C://work/py_exam"))
```
